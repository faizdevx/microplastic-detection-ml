```python
import cv2
import pandas as pd
import numpy as np
import os

rows=[]

for file in image_files:

    img=cv2.imread(file)
    rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    _,thresh=cv2.threshold(
        gray,
        0,
        255,
        cv2.THRESH_BINARY+cv2.THRESH_OTSU
    )

    contours,_=cv2.findContours(
        thresh,
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
    )

    areas=[]
    circularities=[]

    for c in contours:

        area=cv2.contourArea(c)

        if area>10:

            areas.append(area)

            peri=cv2.arcLength(c,True)

            if peri>0:
                circ=4*np.pi*area/(peri*peri)
                circularities.append(circ)

    rows.append({

        "image_name":os.path.basename(file),
        "polymer":"PVC",

        "width":img.shape[1],
        "height":img.shape[0],
        "channels":img.shape[2],

        "particle_count":len(areas),

        "mean_particle_area":
            np.mean(areas) if areas else 0,

        "max_particle_area":
            np.max(areas) if areas else 0,

        "total_particle_area":
            np.sum(areas) if areas else 0,

        "particle_density":
            np.sum(areas)/(img.shape[0]*img.shape[1]),

        "mean_circularity":
            np.mean(circularities) if circularities else 0,

        "mean_red":
            rgb[:,:,0].mean(),

        "mean_green":
            rgb[:,:,1].mean(),

        "mean_blue":
            rgb[:,:,2].mean(),

        "red_std":
            rgb[:,:,0].std(),

        "green_std":
            rgb[:,:,1].std(),

        "blue_std":
            rgb[:,:,2].std()
    })

df=pd.DataFrame(rows)

df.to_csv("PVC_features.csv",index=False)

df.head()
```

```python
import cv2
import pandas as pd
import numpy as np
import os

rows=[]

for file in image_files:

    img = cv2.imread(file)

    if img is None:
        continue

    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    _, thresh = cv2.threshold(
        gray,
        0,
        255,
        cv2.THRESH_BINARY + cv2.THRESH_OTSU
    )

    contours, _ = cv2.findContours(
        thresh,
        cv2.RETR_EXTERNAL,
        cv2.CHAIN_APPROX_SIMPLE
    )

    areas = []
    circularities = []
    aspect_ratios = []

    for c in contours:

        area = cv2.contourArea(c)

        if area > 10:

            areas.append(area)

            peri = cv2.arcLength(c, True)

            if peri > 0:
                circ = 4*np.pi*area/(peri*peri)
                circularities.append(circ)

            x,y,w,h = cv2.boundingRect(c)

            if h > 0:
                aspect_ratios.append(w/h)

    rows.append({

        "image_name":os.path.basename(file),
        "polymer":"NYLON",

        "width":img.shape[1],
        "height":img.shape[0],
        "channels":img.shape[2],

        "particle_count":len(areas),

        "mean_particle_area":
            np.mean(areas) if areas else 0,

        "max_particle_area":
            np.max(areas) if areas else 0,

        "min_particle_area":
            np.min(areas) if areas else 0,

        "total_particle_area":
            np.sum(areas) if areas else 0,

        "particle_density":
            np.sum(areas)/(img.shape[0]*img.shape[1]),

        "mean_circularity":
            np.mean(circularities)
            if circularities else 0,

        "mean_aspect_ratio":
            np.mean(aspect_ratios)
            if aspect_ratios else 0,

        "mean_red":
            rgb[:,:,0].mean(),

        "mean_green":
            rgb[:,:,1].mean(),

        "mean_blue":
            rgb[:,:,2].mean(),

        "red_std":
            rgb[:,:,0].std(),

        "green_std":
            rgb[:,:,1].std(),

        "blue_std":
            rgb[:,:,2].std()
    })

nylon_df = pd.DataFrame(rows)

print(nylon_df.head())

nylon_df.to_csv(
    "NYLON_features.csv",
    index=False
)

print("\nRows:",len(nylon_df))
print("\nColumns:")
print(nylon_df.columns)
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix
)

import matplotlib.pyplot as plt
```

```python
# ===============================
# 2. LOAD DATA
# ===============================

pvc = pd.read_csv("PVC_features.csv")
nylon = pd.read_csv("NYLON_features.csv")


# ===============================
# 3. COMBINE DATASETS
# ===============================

df = pd.concat(
    [pvc, nylon],
    ignore_index=True
)

print("Dataset shape:", df.shape)
print()
print(df['polymer'].value_counts())
```

```python

# ===============================
# 4. REMOVE BAD COLUMNS
# ===============================

# remove columns missing in PVC
drop_cols = [
    'image_name',
    'width',
    'height',
    'channels',
    'min_particle_area',
    'mean_aspect_ratio'
]

df = df.drop(
    columns=drop_cols,
    errors='ignore'
)

print("\nColumns used:")
print(df.columns)
```

```python
X = df.drop(
    columns=['polymer']
)

y = df['polymer']

print("\nFeature matrix:", X.shape)
print("Labels:", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("\nTrain:", X_train.shape)
print("Test :", X_test.shape)
```

```python
rf = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    n_jobs=-1
)

rf.fit(
    X_train,
    y_train
)


# ===============================
# 8. PREDICTION
# ===============================

y_pred = rf.predict(X_test)


# ===============================
# 9. ACCURACY
# ===============================

acc = accuracy_score(
    y_test,
    y_pred
)

print("\nAccuracy:", acc)

```

```python
# ===============================
# 10. CLASSIFICATION REPORT
# ===============================

print("\nClassification Report:\n")

print(
    classification_report(
        y_test,
        y_pred
    )
)


# ===============================
# 11. CONFUSION MATRIX
# ===============================

cm = confusion_matrix(
    y_test,
    y_pred
)

print("\nConfusion Matrix:\n")
print(cm)


# ===============================
# 12. FEATURE IMPORTANCE
# ===============================

importance = pd.DataFrame({

    'Feature': X.columns,
    'Importance': rf.feature_importances_

})

importance = importance.sort_values(
    by='Importance',
    ascending=False
)

print("\nFeature Importance:\n")
print(importance)


# ===============================
# 13. PLOT IMPORTANCE
# ===============================

plt.figure(figsize=(8,6))

plt.barh(
    importance['Feature'],
    importance['Importance']
)

plt.xlabel("Importance")
plt.title("Random Forest Feature Importance")

plt.gca().invert_yaxis()

plt.show()
```

```python
# =====================================
# 1. IMPORTS
# =====================================

import pandas as pd
import numpy as np

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix
)

import matplotlib.pyplot as plt
import seaborn as sns

# =====================================
# 2. LOAD DATA
# =====================================

pvc = pd.read_csv("PVC_features.csv")
nylon = pd.read_csv("NYLON_features.csv")

df = pd.concat(
    [pvc, nylon],
    ignore_index=True
)

print(df.shape)
print(df['polymer'].value_counts())

# =====================================
# 3. REMOVE LEAKAGE FEATURES
# =====================================

drop_cols = [
    'image_name',
    'width',
    'height',
    'channels',
    'min_particle_area',
    'mean_aspect_ratio'
]

df = df.drop(
    columns=drop_cols,
    errors='ignore'
)

# =====================================
# 4. NORMALIZE COLOR FEATURES
# =====================================

total = (
    df['mean_red']
    + df['mean_green']
    + df['mean_blue']
    + 1e-6
)

df['red_ratio'] = df['mean_red']/total
df['green_ratio'] = df['mean_green']/total
df['blue_ratio'] = df['mean_blue']/total

# remove raw brightness
df = df.drop(
    columns=[
        'mean_red',
        'mean_green',
        'mean_blue'
    ]
)

# =====================================
# 5. FEATURES/LABELS
# =====================================

X = df.drop(
    columns=['polymer']
)

y = df['polymer']

print("\nFeatures:")
print(X.columns)

# =====================================
# 6. MODEL
# =====================================

model = Pipeline([

    (
        'scaler',
        StandardScaler()
    ),

    (
        'rf',
        RandomForestClassifier(
            n_estimators=500,
            random_state=42,
            n_jobs=-1
        )
    )
])

# =====================================
# 7. STRATIFIED 5 FOLD
# =====================================

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

scores = []

all_true = []
all_pred = []

for fold,(train,test) in enumerate(
    cv.split(X,y)
):

    X_train = X.iloc[train]
    X_test  = X.iloc[test]

    y_train = y.iloc[train]
    y_test  = y.iloc[test]

    model.fit(
        X_train,
        y_train
    )

    pred = model.predict(
        X_test
    )

    acc = accuracy_score(
        y_test,
        pred
    )

    scores.append(acc)

    all_true.extend(y_test)
    all_pred.extend(pred)

    print(
        f"Fold {fold+1}:",
        round(acc,4)
    )

# =====================================
# 8. RESULTS
# =====================================

print("\n================")
print("FINAL RESULTS")
print("================")

print(
    "Mean Accuracy:",
    round(
        np.mean(scores),
        4
    )
)

print(
    "Std:",
    round(
        np.std(scores),
        4
    )
)

print(
    "\nClassification Report:\n"
)

print(
    classification_report(
        all_true,
        all_pred
    )
)

# =====================================
# 9. CONFUSION MATRIX
# =====================================

cm = confusion_matrix(
    all_true,
    all_pred
)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    xticklabels=['NYLON','PVC'],
    yticklabels=['NYLON','PVC']
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title(
    "Cross Validation Confusion Matrix"
)

plt.show()

# =====================================
# 10. FEATURE IMPORTANCE
# =====================================

rf = model.named_steps['rf']

importance = pd.DataFrame({

    'Feature':
        X.columns,

    'Importance':
        rf.feature_importances_
})

importance = importance.sort_values(
    'Importance',
    ascending=False
)

print("\nFeature Importance:\n")
print(importance)

plt.figure(figsize=(8,6))

plt.barh(
    importance['Feature'],
    importance['Importance']
)

plt.gca().invert_yaxis()

plt.title(
    "Feature Importance"
)

plt.show()
```
